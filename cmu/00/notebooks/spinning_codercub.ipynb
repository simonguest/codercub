# Spinning Codercub

```python
import scene3d
import math
import audio

scene = scene3d.Scene()
scene.set_sky(scene3d.Sky.DEEP_SPACE)

ground = scene.set_ground(20,20)
ground.set_material(scene3d.Material.Chip.CircuitGreen)
ground.set_glossiness(1.0)

box = scene3d.Shapes.Box(width=3, height=3, depth=3)
box.set_position(0,2.5,0)
box.set_texture('/sample_files/codercub.jpg')

scene.add(box)

# Lighting for the scene

scene.ambient.set_brightness(40)

lights = [
  {"x":6, "y":0.5, "z":0, "color":"#ff0000"},
  {"x":0, "y":0.5, "z":6, "color":"#00ff40"},
  {"x":-6, "y":0.5, "z":0, "color":"#5014f5"},
  {"x":0, "y":0.5, "z":-6, "color":"#eeff00"},
]

for light in lights:
  new_light = scene.add_light(light["x"], light["y"], light["z"])
  new_light.set_color(light["color"])
  new_light.set_brightness(60)
  new_light.set_visible(True)

RADIUS = 12
HEIGHT = 5
angle=0
camera_angle = 0

@scene.on_frame
def animate(dt):
  global angle
  global camera_angle
  
  angle += dt * 60  # 60 degrees per second → 1 degree per frame at 60fps
  camera_angle += dt * 0.5
  
  box.set_rotation(x=angle, y=-angle, z=-angle)

  # Move the camera
  x = RADIUS * math.sin(camera_angle)
  z = RADIUS * math.cos(camera_angle)
  scene.camera.set_position(x, HEIGHT, z).look_at(0, 0, 0)

await audio.play_async("/sample_files/space_ambience.wav")

scene.run()
```
