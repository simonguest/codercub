# Variables

When you write a Python program, you need a way to keep track of the things you create.
A **variable** is a name you attach to a value — a label that lets you find it and work with it later.

In this lesson you'll create a 3D scene and a sphere, and see exactly how variables make that possible.

## What Is a Variable?

Think of a variable like a label on a box. The label isn't the thing inside — it's just a way to find it.

In Python, you create a variable using `=`:

```python
sphere = scene3d.Shapes.Sphere(5)
```

The right side creates a sphere. The `=` attaches the label `sphere` to it.
From that point on, whenever you write `sphere` in your code, Python knows exactly which sphere you mean.

> The `=` in Python does **not** mean "equal to" — it means **"store this value with this name"**.

## A Scene and a Sphere

Run the cell below. A sphere appears in a 3D scene!

> Hint: Hold the left mouse button to rotate the scene. Use the scroll wheel to zoom.

Two variables make this happen — `scene` and `sphere`.
Look for them in the code before you run it.

````python
import scene3d

scene = scene3d.Scene()
sphere = scene3d.Shapes.Sphere(5)
scene.add(sphere)
````

## How the Variables Work

Two lines in that code create variables:

```python
scene  = scene3d.Scene()
sphere = scene3d.Shapes.Sphere(5)
```

| Variable | What it holds |
|----------|---------------|
| `scene` | The 3D scene — the world where everything lives |
| `sphere` | The sphere shape — an object inside that world |

Once a variable exists, you use its name to work with what it holds:

| Line | What happens |
|------|--------------|
| `scene.add(sphere)` | Uses **both** variables — places the sphere into the scene |
