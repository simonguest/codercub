# The 3D Coordinate System

Every 3D scene is built on an invisible grid that stretches in three directions at once.
Once you understand that grid, you can place any object exactly where you want it.

```python
import scene3d

def add_axes(scene):
    L = 3.0
    R = 0.06
    CD = 0.28
    CH = 0.45

    # X Axis — Red
    shaft = scene3d.Shapes.Cylinder(diameter=R * 2, height=L * 2, tessellation=8)
    shaft.set_rotation(z=90)
    shaft.set_color("#e53935")
    scene.add(shaft)

    cone = scene3d.Shapes.Cone(diameter=CD, height=CH, tessellation=12)
    cone.set_position(L + CH / 2, 0, 0)
    cone.set_rotation(z=-90)
    cone.set_color("#e53935")
    scene.add(cone)

    cone = scene3d.Shapes.Cone(diameter=CD, height=CH, tessellation=12)
    cone.set_position(-L - CH / 2, 0, 0)
    cone.set_rotation(z=90)
    cone.set_color("#e53935")
    scene.add(cone)

    lbl = scene3d.Shapes.Text("+X", 0.5, 0.08, 8)
    lbl.set_position(L + CH + 0.1, -0.25, 0)
    lbl.set_color("#e53935")
    scene.add(lbl)

    lbl = scene3d.Shapes.Text("-X", 0.5, 0.08, 8)
    lbl.set_position(-L - CH - 0.7, -0.25, 0)
    lbl.set_color("#e53935")
    scene.add(lbl)

    # Y Axis — Green
    shaft = scene3d.Shapes.Cylinder(diameter=R * 2, height=L * 2, tessellation=8)
    shaft.set_color("#43a047")
    scene.add(shaft)

    cone = scene3d.Shapes.Cone(diameter=CD, height=CH, tessellation=12)
    cone.set_position(0, L + CH / 2, 0)
    cone.set_color("#43a047")
    scene.add(cone)

    cone = scene3d.Shapes.Cone(diameter=CD, height=CH, tessellation=12)
    cone.set_position(0, -L - CH / 2, 0)
    cone.set_rotation(x=180)
    cone.set_color("#43a047")
    scene.add(cone)

    lbl = scene3d.Shapes.Text("+Y", 0.5, 0.08, 8)
    lbl.set_position(-0.25, L + CH + 0.1, 0)
    lbl.set_color("#43a047")
    scene.add(lbl)

    lbl = scene3d.Shapes.Text("-Y", 0.5, 0.08, 8)
    lbl.set_position(-0.25, -L - CH - 0.6, 0)
    lbl.set_color("#43a047")
    scene.add(lbl)

    # Z Axis — Blue
    shaft = scene3d.Shapes.Cylinder(diameter=R * 2, height=L * 2, tessellation=8)
    shaft.set_rotation(x=90)
    shaft.set_color("#1e88e5")
    scene.add(shaft)

    cone = scene3d.Shapes.Cone(diameter=CD, height=CH, tessellation=12)
    cone.set_position(0, 0, L + CH / 2)
    cone.set_rotation(x=90)
    cone.set_color("#1e88e5")
    scene.add(cone)

    cone = scene3d.Shapes.Cone(diameter=CD, height=CH, tessellation=12)
    cone.set_position(0, 0, -L - CH / 2)
    cone.set_rotation(x=-90)
    cone.set_color("#1e88e5")
    scene.add(cone)

    lbl = scene3d.Shapes.Text("+Z", 0.5, 0.08, 8)
    lbl.set_position(-0.25, -0.25, L + CH + 0.1)
    lbl.set_rotation(y=180)
    lbl.set_color("#1e88e5")
    scene.add(lbl)

    lbl = scene3d.Shapes.Text("-Z", 0.5, 0.08, 8)
    lbl.set_position(-0.25, -0.25, -L - CH - 0.7)
    lbl.set_color("#1e88e5")
    scene.add(lbl)
```

## Three Directions, Three Axes

In a 3D scene, every location is described by three numbers: `(X, Y, Z)`.

| Coordinate | Direction |
|---|---|
| **X** | Left and right |
| **Y** | Up and down |
| **Z** | Forward and backward |

The point `(0, 0, 0)` is called the **origin** — the centre of everything.
Every other position is measured as a distance away from that centre point.

## Visualising the Axes

Run the cell below to see all three axes in a 3D scene.
Each axis is a shaft with an arrowhead at both ends — one pointing in the positive direction, one in the negative.

> **Tip:** Hold the left mouse button to orbit the scene. Use the scroll wheel to zoom in and out.

```python
import scene3d

scene = scene3d.Scene()
scene.set_sky("#0d1b2a")

add_axes(scene)
scene.camera.set_position(2, 2, -10).look_at(0, 0, 0)
```

## Reading the Diagram

The axes follow a colour convention used by 3D artists and game engines everywhere:

| Axis | Colour | Direction |
|---|---|---|
| X | Red | Left and right |
| Y | Green | Up and down |
| Z | Blue | Into and out of the scene |

A handy memory trick: **RGB = XYZ**.

The label at each end shows whether you are going in the **positive** (`+`) or **negative** (`-`) direction.
Positive X is to the right. Negative X is to the left. Positive Y is up. And so on.

When you write `set_position(2, 0, 0)` in your code, you are placing an object
**2 units to the right** of the origin, along the red X axis.

## Place the Ball

Use the sliders below to move a yellow ball anywhere in the scene.
Watch how changing each coordinate moves the ball in a different direction.

- Can you move the ball to the very top of the Y axis?
- Can you place it exactly at the origin `(0, 0, 0)`?
- What happens when a coordinate goes negative?

```python
import scene3d

X_POSITION = 3 #@param {type:"slider", min:-3, max:3, step:0.5}
Y_POSITION = 0 #@param {type:"slider", min:-3, max:3, step:0.5}
Z_POSITION = 0 #@param {type:"slider", min:-3, max:3, step:0.5}

scene = scene3d.Scene()
scene.set_sky("#0d1b2a")
add_axes(scene)

ball = scene3d.Shapes.Sphere(diameter=0.6, segments=16)
ball.set_position(X_POSITION, Y_POSITION, Z_POSITION)
ball.set_color("#ffeb3b")
scene.add(ball)

print(f"Ball is at ({X_POSITION}, {Y_POSITION}, {Z_POSITION})")
scene.camera.set_position(2, 2, -10).look_at(0, 0, 0)
```
