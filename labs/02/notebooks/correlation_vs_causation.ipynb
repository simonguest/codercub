## Step 1: The Data

A city health department tracked three things every month for a year:

- **`temp_c`** — average temperature in °C
- **`ice_cream_sales`** — average weekly ice cream sales (units)
- **`sunburn_cases`** — average weekly sunburn cases reported at local clinics

Run the cell below to load the data.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
  'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
  'temp_c': [3,  4,  9, 14, 19, 23, 26, 25, 20, 14,  8,  4],
  'ice_cream_sales': [120, 145, 290, 440, 690, 900, 1060, 990, 720, 400, 215, 135],
  'sunburn_cases': [8,  11,  28,  52,  95, 148, 182, 167, 102,  45,  20,  10]
}

df = pd.DataFrame(data)
print(df)
```

## Step 2: A Suspicious Pattern

Let's plot ice cream sales against sunburn cases to see if there's a relationship.

**Before you run:** what do you expect to see — an upward slope, a downward slope, or random scatter?

Write your prediction: do you think ice cream sales and sunburn cases will be correlated? Why or why not?

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(df['ice_cream_sales'], df['sunburn_cases'],
           color='steelblue', s=80, alpha=0.85)

for _, row in df.iterrows():
  ax.annotate(row['month'],
    (row['ice_cream_sales'], row['sunburn_cases']),
    textcoords='offset points', xytext=(6, 4), fontsize=8)

corr = df['ice_cream_sales'].corr(df['sunburn_cases'])
ax.text(0.04, 0.92, f'Correlation: {corr:.2f}', transform=ax.transAxes,
        fontsize=11, bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.85))

ax.set_xlabel('Ice Cream Sales (units/week)', fontsize=12)
ax.set_ylabel('Sunburn Cases (per week)', fontsize=12)
ax.set_title('Ice Cream Sales vs. Sunburn Cases', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

The correlation is very strong. Does that mean buying ice cream causes sunburn? Write your reaction — does this result make sense to you, or does something feel off?

## Step 3: Investigating Temperature

The dataset also contains `temp_c`. A hunch: maybe temperature is driving *both* ice cream sales and sunburn cases at the same time — and ice cream and sunburn are just along for the ride.

Let's test that by plotting temperature against each variable separately.

### Temperature vs. Ice Cream Sales

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(df['temp_c'], df['ice_cream_sales'],
           color='coral', s=80, alpha=0.85)

for _, row in df.iterrows():
    ax.annotate(row['month'],
                (row['temp_c'], row['ice_cream_sales']),
                textcoords='offset points', xytext=(6, 4), fontsize=8)

corr = df['temp_c'].corr(df['ice_cream_sales'])
ax.text(0.04, 0.92, f'Correlation: {corr:.2f}', transform=ax.transAxes,
        fontsize=11, bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.85))

ax.set_xlabel('Temperature (°C)', fontsize=12)
ax.set_ylabel('Ice Cream Sales (units/week)', fontsize=12)
ax.set_title('Temperature vs. Ice Cream Sales', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

### Temperature vs. Sunburn Cases

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(df['temp_c'], df['sunburn_cases'],
           color='goldenrod', s=80, alpha=0.85)

for _, row in df.iterrows():
    ax.annotate(row['month'],
                (row['temp_c'], row['sunburn_cases']),
                textcoords='offset points', xytext=(6, 4), fontsize=8)

corr = df['temp_c'].corr(df['sunburn_cases'])
ax.text(0.04, 0.92, f'Correlation: {corr:.2f}', transform=ax.transAxes,
        fontsize=11, bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.85))

ax.set_xlabel('Temperature (°C)', fontsize=12)
ax.set_ylabel('Sunburn Cases (per week)', fontsize=12)
ax.set_title('Temperature vs. Sunburn Cases', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

What correlation values did you get for temperature vs. ice cream, and temperature vs. sunburn? How do they compare to the ice cream vs. sunburn value from Step 2?

## Step 4: The Reveal — A Hidden Third Variable

Here's what the data is actually telling us:

- High temperature → more ice cream sales (hot days = more ice cream)
- High temperature → more sunburn cases (hot days = more sun exposure)
- Ice cream and sunburn both follow temperature — so they end up correlated with *each other*

**Temperature is a confounding variable** — a hidden third factor that drives both variables at once and makes them look connected when they aren't.

Ice cream doesn't cause sunburn. The relationship is an illusion created by the season.

This is one of the most important ideas in data science:

> **Correlation does not imply causation.**

Two variables can move together perfectly without either one causing the other. Before drawing conclusions from a correlation, always ask: *could a third variable explain both?*
