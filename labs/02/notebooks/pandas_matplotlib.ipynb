## Part 1: Hello, Pandas!

**Pandas** is a Python library for working with structured data — like spreadsheets or tables.

The key building block is a **DataFrame** — think of it as a table with rows and columns, just like a spreadsheet.

Let's create a DataFrame using average monthly temperatures in New York City:

```python
import pandas as pd

data = {
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'avg_temp_c': [0, 1, 6, 12, 18, 23, 26, 25, 21, 15, 9, 3]
}

df = pd.DataFrame(data)
df
```

## What is a Series?

A **Series** is a single column of data — like one column pulled from your DataFrame.

You can access a column by name using square brackets `[]`. Let's look at just the temperature column:

```python
temperatures = df['avg_temp_c']
print(temperatures)
```

## Exploring Your Data

- `df.shape` — tells you how many rows and columns you have

## Part 2: Making Charts with Matplotlib

Numbers in a table are useful, but charts make patterns *much* easier to see!

**Matplotlib** is a Python library for creating charts and visualizations.

The most common chart types are:
- **Line chart** — great for showing change over time
- **Bar chart** — great for comparing values side by side

Let's plot our temperature data as a line chart:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['month'], df['avg_temp_c'], marker='o', color='steelblue', linewidth=2)
plt.title('Average Monthly Temperature — New York City')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Now let's show the same data as a **bar chart**. Months below freezing (0°C) will appear in a different colour:

```python
colors = ['steelblue' if t >= 0 else 'slateblue' for t in df['avg_temp_c']]

plt.figure(figsize=(10, 5))
plt.bar(df['month'], df['avg_temp_c'], color=colors)
plt.title('Average Monthly Temperature — New York City')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.axhline(y=0, color='black', linewidth=0.8)
plt.show()
```

## Explore Different Cities

Use the dropdown below to pick a different city and see how the temperature pattern changes!

```python
CITY = "Sydney" #@param ["New York", "Sydney", "London", "Nairobi"]

city_data = {
    "New York":  [0,  1,  6, 12, 18, 23, 26, 25, 21, 15,  9,  3],
    "Sydney":    [26, 26, 24, 21, 18, 15, 14, 15, 17, 20, 22, 25],
    "London":    [6,  6,  9, 12, 15, 18, 20, 20, 17, 13,  9,  6],
    "Nairobi":   [22, 23, 23, 22, 21, 19, 18, 19, 21, 22, 21, 21]
}

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temps = city_data[CITY]

plt.figure(figsize=(10, 5))
plt.plot(months, temps, marker='o', color='coral', linewidth=2)
plt.title(f'Average Monthly Temperature — {CITY}')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

What differences did you notice between cities? Which city has the most stable temperature year-round? Why do you think that is?
