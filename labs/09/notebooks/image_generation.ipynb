# Image Generation

In this notebook, we use an image model to generate an image from a text prompt. This is known as a text-to-image model and uses the diffusion process that we covered in class.

```python
import openai
import os

client = openai.OpenAI(
  base_url='https://openrouter.ai/api/v1',
   api_key=os.environ["OPENROUTER_API_KEY"]
)

MODEL = "black-forest-labs/flux.2-klein-4b" #@param ["black-forest-labs/flux.2-klein-4b"]

response = client.chat.completions.create(
  model=MODEL,
  messages=[
    {"role": "user", "content": "A Mongolian Bankhar dog running across the steppe at sunset"},
  ],
  extra_body={
    "modalities":["image"]
  }
)
```

Now we need to extract the image from the result...

```python
image_url = response.choices[0].message.images[0]["image_url"]["url"]
print(image_url[:500])
```

And display the image...

```python
import base64
from PIL import Image
from io import BytesIO
from IPython.display import display

# Strip the data URI prefix if present
if image_url.startswith("data:"):
    image_url = image_url.split(",", 1)[1]

# Decode and display
image_data = base64.b64decode(image_url)
image = Image.open(BytesIO(image_data))
display(image)
```
